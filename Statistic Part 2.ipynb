Theory Questions

1.	What is hypothesis testing in statistics?
Ans. Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps you determine whether there is enough evidence in a sample to support a specific claim or hypothesis about a population.
Example:
Suppose a researcher claims that a new drug lowers blood pressure more effectively than the old one.
H₀: The new drug is not more effective than the old one.
H₁: The new drug is more effective than the old one.
Collect sample data, perform a statistical test (e.g., t-test), compute the p-value, and decide whether to reject H₀.

2.	What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans. 1. Null Hypothesis (H₀):
Definition:
The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population.
Purpose:
It serves as the default assumption or baseline that we test against.
Example:
Suppose we want to test whether a new teaching method is more effective than the traditional one.
H₀: The new method is not more effective than the traditional method.
Or, formally: H₀: μ₁ = μ₂ (where μ represents the population mean scores of the two methods)
2. Alternative Hypothesis (H₁ or Ha):
Definition:
The alternative hypothesis is a statement that contradicts the null. It claims that there is an effect, a difference, or a relationship.
Purpose:
This is what you are trying to find evidence for.
Example:
Continuing the same scenario:
H₁: The new method is more effective than the traditional one.
Or, formally: H₁: μ₁ ≠ μ₂ (or μ₁ > μ₂ depending on the direction of the hypothesis)


3.	What is the significance level in hypothesis testing, and why is it important?
Ans. The significance level, often denoted by α (alpha), is a threshold used to decide whether the results of a hypothesis test are statistically significant.

α = 0.05 (most common) → 5% chance of making a Type I error
α = 0.01 → more strict, 1% chance
α = 0.10 → more lenient, 10% chance


4.	What does a P-value represent in hypothesis testing?
Ans. The p-value is a central concept in hypothesis testing. It helps you decide whether the evidence in your sample is strong enough to reject the null hypothesis (H₀).

5.	How do you interpret the P-value in hypothesis testing?
Ans. 
if p ≤ 0.05 then Reject H₀
if p > 0.05 then Do not reject H₀

6.	What are Type 1 and Type 2 errors in hypothesis testing?
Ans.
Type I Error (False Positive):
Definition: Rejecting the null hypothesis (H₀) when it is actually true.

Interpretation: You conclude there's an effect or difference, but there actually isn't.

Probability of making this error: Equal to the significance level (α), often set at 0.05.

Example: Saying a new drug works better than the current one, when in reality, it doesn't.

Type II Error (False Negative):
Definition: Failing to reject the null hypothesis when it is actually false.

Interpretation: You conclude there's no effect or difference, but in reality, there is.

Probability of making this error: 
Denoted by β (beta).

Power of a test = 1 − β (the probability of correctly detecting a true effect).

Example: Saying a new teaching method doesn't improve scores, when in fact, it does.

7.	What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Ans. The choice between a one-tailed and a two-tailed test depends on your research question and the direction of the effect you want to detect.
 1. One-Tailed Test:
Definition: Tests whether a parameter is greater than or less than a specific value, but not both.

Used When: We have a directional hypothesis — we're only interested in one direction of change.

Example:
To test if a new drug increases recovery rate.

H₀: Recovery rate ≤ 70%

H₁: Recovery rate > 70% → Right-tailed test

Advantage: More statistical power to detect an effect in the specified direction.

Risk: If the effect is in the opposite direction, the test won’t detect it.

2. Two-Tailed Test:
Definition: Tests whether a parameter is different (either higher or lower) than a specific value.

Used When: You have a non-directional hypothesis — you're open to effects in either direction.

Example: Testing whether a new teaching method is different (better or worse) than the current one.

H₀: Mean = 75

H₁: Mean ≠ 75

Advantage: Detects effects in both directions.

Downside: Requires a more extreme test statistic to reach significance (since α is split between both tails).

8.	What is the Z-test, and when is it used in hypothesis testing?
Ans.
A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means, or between two sample means, when the population variance is known or the sample size is large.
Use a Z-test when all of the following conditions are met:
The sample size is large (typically n ≥ 30).
The population standard deviation (σ) is known.
The data is approximately normally distributed (or the sample size is large enough for the Central Limit Theorem to apply).

9.	How do you calculate the Z-score, and what does it represent in hypothesis testing?
Ans. A Z-score (or standard score) measures how many standard deviations a data point or sample statistic is from the population mean. In hypothesis testing, it helps you determine whether the observed sample result is statistically significant.
Z= (x−μ)/σ/sqrt(n)
 
Where:
x= sample mean
μ = population mean (under H₀)
σ = population standard deviation
n = sample size

10.	What is the T-distribution, and when should it be used instead of the normal distribution?
Ans. The t-distribution is a probability distribution used in hypothesis testing when the sample size is small and/or the population standard deviation is unknown. It is similar in shape to the normal distribution but has heavier tails, meaning it accounts for more variability.
Use the T-distribution when:
The population standard deviation (σ) is unknown
The sample size is small (typically n < 30)
The data is approximately normally distributed

11.	What is the difference between a Z-test and a T-test?

Both Z-tests and T-tests are used to test hypotheses about population means or proportions — but they are used under different conditions based on what is known about the population and sample.
Large sample size and population SD is known	    Z-test
Small sample size and population SD is unknown	    T-test
Testing sample proportion (with large sample)	    Z-test
Comparing two small sample means with unknown SDs	T-test

19.	What is the F-distribution, and when is it used in hypothesis testing?
Ans. The F-distribution is a continuous probability distribution that arises primarily in the context of comparing variances of two populations. It is the distribution of the ratio of two scaled chi-square variables, each divided by their respective degrees of freedom.
When Is the F-Distribution Used?
Testing Equality of Variances: To test if two populations have the same variance (using the F-test for equality of variances).
Analysis of Variance (ANOVA): To compare means across three or more groups by analyzing variance between groups versus within groups.
Regression Analysis: To test the overall significance of a regression model.

20.	What is an ANOVA test, and what are its assumptions?
Ans. ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to see if at least one group mean is significantly different from the others.
Assumptions of ANOVA:
Independence: Observations are independent of each other.
Normality: The data in each group are approximately normally distributed.
Homogeneity of variances (Homoscedasticity): Variances across groups are roughly equal.
Scale of Measurement: The dependent variable is continuous (interval or ratio scale).

21.	What are the different types of ANOVA tests?
Ans. 1. One-Way ANOVA
Purpose: Compares means of three or more independent groups based on one factor (independent variable).
Example: Comparing exam scores across three different teaching methods.
Key point: Only one factor (independent variable).

2. Two-Way ANOVA (Factorial ANOVA)
Purpose: Tests the effect of two independent variables (factors) on a dependent variable, including their interaction effect.
Example: Studying the impact of teaching method (Factor 1) and study time (Factor 2) on exam scores.
Key point: Evaluates main effects of each factor and interaction between them.

3. Repeated Measures ANOVA
Purpose: Used when the same subjects are measured multiple times under different conditions or over time.
Example: Measuring blood pressure of patients before, during, and after medication.
Key point: Accounts for correlations between repeated observations on the same subjects.

4. Mixed-Design ANOVA (Split-Plot ANOVA)
Purpose: Combines both between-subjects factors (different groups) and within-subjects factors (repeated measures).
Example: Testing the effect of a drug (treatment vs placebo) across several time points in the same subjects.
Key point: Handles mixed types of data designs.

5. MANOVA (Multivariate ANOVA)
Purpose: Tests differences on multiple dependent variables simultaneously across groups.
Example: Comparing groups based on multiple test scores (math, reading, science).
Key point: Extends ANOVA when there are multiple outcomes.

22.	What is the F-test, and how does it relate to hypothesis testing?
Ans. The F-test is a statistical test that uses the F-distribution to compare variances or test hypotheses about the equality of group means in certain models.
Purpose in Hypothesis Testing:
Compare variances: Test if two populations have equal variances.
Compare group means (ANOVA): Test if the means across three or more groups are equal.
Test overall significance in regression: Check if a regression model explains a significant amount of variation.

Null hypothesis (H₀): No difference in group means (or variances are equal).
Alternative hypothesis (H₁): At least one group mean is different (or variances differ).

2.	Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution parameters
low = 1   # inclusive lower bound
high = 7  # exclusive upper bound, so this covers values 1 to 6

# Create the discrete uniform distribution
dist = randint(low, high)

# Values for the PMF (all integers in the range)
x = np.arange(low, high)

# Calculate PMF values
pmf = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

3.	Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    - x : int (0 or 1) - the outcome
    - p : float (0 <= p <= 1) - probability of success

    Returns:
    - Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7
print(f"P(X=1) = {bernoulli_pdf(1, p)}")
print(f"P(X=0) = {bernoulli_pdf(0, p)}")
```

9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):
    """
    Calculate Z-scores of the data and plot them.

    Parameters:
    - data: array-like, the dataset

    Returns:
    - z_scores: numpy array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage:
sample_data = [10, 12, 23, 23, 16, 23, 21, 16]
z = plot_z_scores(sample_data)
print("Z-scores:", z)
```

15.Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05, tail='two-tailed'):
    """
    Perform hypothesis testing using Z-statistics.
    
    Parameters:
    - sample_data: list or array of sample values
    - population_mean: hypothesized population mean (μ₀)
    - population_std: known population standard deviation (σ)
    - alpha: significance level (default 0.05)
    - tail: 'two-tailed', 'left-tailed', or 'right-tailed'
    """
    sample = np.array(sample_data)
    sample_mean = np.mean(sample)
    n = len(sample)
    se = population_std / np.sqrt(n)
    z = (sample_mean - population_mean) / se

    if tail == 'two-tailed':
        p_value = 2 * (1 - norm.cdf(abs(z)))
    elif tail == 'left-tailed':
        p_value = norm.cdf(z)
    elif tail == 'right-tailed':
        p_value = 1 - norm.cdf(z)
    else:
        raise ValueError("tail must be 'two-tailed', 'left-tailed', or 'right-tailed'")

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("➡️ Reject the null hypothesis (significant difference).")
    else:
        print("➡️ Fail to reject the null hypothesis (no significant difference).")

# Example usage
sample_data = [72, 70, 68, 74, 71, 69, 73, 75, 70, 68]
population_mean = 70
population_std = 3  # known population standard deviation

z_test(sample_data, population_mean, population_std, alpha=0.05, tail='two-tailed')
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.
    
    Parameters:
    - data: list or numpy array of numerical values
    - confidence: confidence level (default = 0.95 for 95%)
    
    Returns:
    - (lower_bound, upper_bound)
    """
    sample = np.array(data)
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # Two-tailed t critical value

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident the true population mean lies within this interval.")
    
    return (lower, upper)

# Example usage
sample_data = [72, 70, 68, 74, 71, 69, 73, 75, 70, 68]
confidence_interval(sample_data, confidence=0.95)
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
import numpy as np
from scipy.stats import norm
import math

def one_sample_z_test(sample_data, population_mean, population_std_dev, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.
    
    Parameters:
    - sample_data: array-like of sample values
    - population_mean: known population mean
    - population_std_dev: known population standard deviation
    - alpha: significance level (default 0.05)
    - alternative: 'two-sided', 'greater', or 'less'
    
    Returns:
    - z_score, p_value, interpretation string
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std_dev / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / standard_error

    # Compute p-value based on alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis."
    else:
        result = "Fail to reject the null hypothesis."

    return z_score, p_value, result

# --- Example usage ---
np.random.seed(1)
population_mean = 100
population_std_dev = 15
sample = np.random.normal(loc=102, scale=population_std_dev, size=50)

z, p, interpretation = one_sample_z_test(sample, population_mean, population_std_dev, alpha=0.05, alternative='two-sided')

# --- Output ---
print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p:.4f}")
print(f"Result: {interpretation}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

# --- Step 1: Define Parameters ---
population_mean = 100
population_std_dev = 15
sample_size = 50
alpha = 0.05

# Simulate sample data
np.random.seed(0)
sample_data = np.random.normal(loc=103, scale=population_std_dev, size=sample_size)
sample_mean = np.mean(sample_data)

# --- Step 2: Calculate Z-score ---
standard_error = population_std_dev / math.sqrt(sample_size)
z_score = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # two-tailed

# --- Step 3: Decision Boundaries ---
z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed critical value

# --- Step 4: Print Results ---
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{z_critical:.4f}")

if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis.")
else:
    print("Result: Fail to reject the null hypothesis.")

# --- Step 5: Visualization ---
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4, label='Rejection Region (right)')

# Draw z-score line
plt.axvline(z_score, color='black', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='gray', linestyle=':', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='gray', linestyle=':')

# Labels and legend
plt.title('Two-Tailed Z-Test: Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05, one_tailed=False):
    """
    Visualize Type I and Type II errors for a one-sample Z-test.
    
    Parameters:
    - mu_0: population mean under H0
    - mu_1: true mean under H1
    - sigma: known population standard deviation
    - n: sample size
    - alpha: significance level
    - one_tailed: if True, perform a one-tailed test (right-tailed)
    """
    se = sigma / np.sqrt(n)
    
    # Calculate critical value(s)
    if one_tailed:
        z_crit = norm.ppf(1 - alpha)
        critical_value = mu_0 + z_crit * se
        beta = norm.cdf((critical_value - mu_1) / se)
    else:
        z_crit = norm.ppf(1 - alpha/2)
        critical_low = mu_0 - z_crit * se
        critical_high = mu_0 + z_crit * se
        beta = norm.cdf(critical_high - mu_1) / se - norm.cdf((critical_low - mu_1) / se)

    # Plot range
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y0 = norm.pdf(x, mu_0, se)
    y1 = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀: μ = {:.1f}'.format(mu_0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {:.1f}'.format(mu_1), color='green')

    if one_tailed:
        # Shade Type I error region (right tail of H₀)
        plt.fill_between(x, y0, where=(x >= critical_value), color='red', alpha=0.5, label='Type I Error (α)')
        # Shade Type II error region (left of critical under H₁)
        plt.fill_between(x, y1, where=(x < critical_value), color='orange', alpha=0.5, label='Type II Error (β)')
        plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
    else:
        # Shade Type I error regions (tails of H₀)
        plt.fill_between(x, y0, where=(x <= critical_low), color='red', alpha=0.5)
        plt.fill_between(x, y0, where=(x >= critical_high), color='red', alpha=0.5, label='Type I Error (α)')
        # Shade Type II error region (middle of H₁)
        plt.fill_between(x, y1, where=(x > critical_low) & (x < critical_high), color='orange', alpha=0.5, label='Type II Error (β)')
        plt.axvline(critical_low, color='black', linestyle='--')
        plt.axvline(critical_high, color='black', linestyle='--', label=f'Critical Values = {critical_low:.2f}, {critical_high:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Significance Level (α): {alpha:.3f}")
    print(f"Type II Error Probability (β): {beta:.3f}")
    print(f"Power of the test (1 - β): {1 - beta:.3f}")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean using the t-distribution.

    Parameters:
    - data: list or array of sample data
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): confidence interval for the mean
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error with sample std dev
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper, mean, margin_of_error
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean using the t-distribution.

    Parameters:
    - data: list or numpy array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - margin of error
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    se = sample_std / np.sqrt(n)       # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    moe = t_crit * se
    return moe
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.
    
    Parameters:
    - prior: P(H) probability of hypothesis before seeing evidence
    - likelihood: P(E|H) probability of evidence given hypothesis
    - evidence: P(E) total probability of evidence
    
    Returns:
    - posterior: P(H|E) updated probability after observing evidence
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Given data
P_disease = 0.01          # Prior: disease prevalence
P_no_disease = 0.99       # Complement

P_pos_given_disease = 0.99   # True positive rate (sensitivity)
P_pos_given_no_disease = 0.01 # False positive rate (1 - specificity)

# Total probability of testing positive (evidence)
P_pos = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Calculate posterior probability: P(disease | positive test)
posterior = bayes_theorem(P_disease, P_pos_given_disease, P_pos)

print(f"Probability of having disease given positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.
    
    Parameters:
    - observed: 2D numpy array of observed counts
    
    Returns:
    - expected: 2D numpy array of expected counts
    """
    # Row totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    # Column totals
    col_totals = observed.sum(axis=0).reshape(1, -1)
    # Grand total
    grand_total = observed.sum()
    
    # Calculate expected frequencies
    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example observed data
observed = np.array([
    [30, 10, 20],
    [20, 25, 15]
])

expected = calculate_expected_frequencies(observed)
print("Observed Frequencies:\n", observed)
print("Expected Frequencies:\n", expected)
```
